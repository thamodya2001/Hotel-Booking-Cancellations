```python
y_train = pd.read_excel("F:\\UOC\\3rd Year\\2nd Sem\\ST 3082 - Statistical Learning I\\Group 4\\Project III\\booking cancellation\\y_train.xlsx")
y_train.head()
```

```python
y_test = pd.read_excel("F:\\UOC\\3rd Year\\2nd Sem\\ST 3082 - Statistical Learning I\\Group 4\\Project III\\booking cancellation\\y_test.xlsx")
y_test.head()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, accuracy_score
```

```python
# Month mapping (custom order)
month_map = {
    'January': 1, 'February': 2, 'March': 3, 'April': 4,
    'May': 5, 'June': 6, 'July': 7, 'August': 8,
    'September': 9, 'October': 10, 'November': 11, 'December': 12
}

# Apply the mapping
X_train['arrival_date_month'] = X_train['arrival_date_month'].map(month_map)
```

```python
label_encoders = {}

for col in X_train.select_dtypes(include='object').columns:
    le = LabelEncoder()
    X_train[col] = le.fit_transform(X_train[col])
    label_encoders[col] = le
```

```python
X_train.head()
```

```python
# Initialize and fit the model
clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
# Evaluate the model (on training data for this example)
y_trpred = clf.predict(X_train)

print("Model Accuracy:", accuracy_score(y_train, y_trpred))
print("\nClassification Report:")
print(classification_report(y_train, y_trpred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_train, y_trpred))
```

```python
# Encode the categorical columns in X_test using the same label encoders
# Apply the mapping
X_test['arrival_date_month'] = X_test['arrival_date_month'].map(month_map)

for col, le in label_encoders.items():
    if col in X_test:
        X_test[col] = le.transform(X_test[col])

# Make predictions
y_pred = clf.predict(X_test)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
importances = pd.Series(clf.feature_importances_, index=X_train.columns)
print(importances.sort_values(ascending=False).head(10))
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier

param_grid = {
    'max_depth': [5, 10, 15, 20, None],
    'min_samples_split': [2, 10, 20],
    'min_samples_leaf': [1, 5, 10],
    'criterion': ['gini', 'entropy']
}

grid_search = GridSearchCV(DecisionTreeClassifier(), param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)

print("Best Parameters:", grid_search.best_params_)
best_tree = grid_search.best_estimator_
```

```python
# Evaluate the model (on training data for this example)
y_trpred = best_tree.predict(X_train)

print("Model Accuracy:", accuracy_score(y_train, y_trpred))
print("\nClassification Report:")
print(classification_report(y_train, y_trpred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_train, y_trpred))
```

```python
# Make predictions
y_pred = best_tree.predict(X_test)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
